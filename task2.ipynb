```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/princess.png')
cv2_imshow(img)
```

```python
img1= cv2.imread('/content/drive/MyDrive/Colab Notebooks/princess.png') 
blur_image = cv2.blur(img1,(4,4))
print("diminished dimensions",blur_image.shape) 
cv2_imshow(blur_image)
```

```python
img2 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/princess1.png')
scale_percent = 60
width = int(img2.shape[1] * scale_percent / 100)
height = int(img2.shape[0] * scale_percent / 100)
dim = (width, height)
resized_blur_image = cv2.resize(img2,dim)
cv2_imshow(resized_blur_image)
```

```python
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/princess2.png')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
grayblur =cv2.blur(gray_image,(4,4))
cv2_imshow(grayblur)
```

```python
pencil_sketch =cv2.divide(gray_image,grayblur,scale=220.0)
cv2_imshow(pencil_sketch)
```
